```python
import networkx as nx # for network and graph analysis
import math
```

```python
routers = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
links = [('A', 'B') , ('A', 'C') , ('A', 'D') , ('B', 'D') , ('C', 'D') , ('C', 'E') ,
         ('D', 'E') , ('D', 'F') , ('E', 'F') , ('E', 'G') , ('F', 'G') ]
link_costs = {('A', 'B') : 2 , ('A', 'C') : 2 , ('A', 'D') : 2 , ('B', 'D') : 3 , ('C', 'D') : 1 ,
              ('C', 'E') : 3 , ('D', 'E') : 1 , ('D', 'F') : 4 , ('E', 'F') : 2 , ('E', 'G') : 3 , ('F', 'G') : 1}
```

```python
G = nx.Graph() # G is the variable holding our network object .
G.add_nodes_from(routers)
for link in links :
    l1 , l2 = link
    cost = link_costs.get (link , 0)
    G.add_edge (l1, l2, weight = cost )
```

```python
def draw_network(G):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels = True, font_weight = 'bold')
    labels = nx.get_edge_attributes(G , 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels = labels)
```

```python
draw_network(G)
```

```python
def dijkstra(G, source):
    shortest_path_lengths, shortest_paths = nx.single_source_dijkstra(G, source, weight='weight')

    # print the results for each destination
    for destination, path in shortest_paths.items():
        print(f"Shortest path from {source} to {destination}: Length = {shortest_path_lengths[destination]}, Path = {path}")
```

```python
dijkstra(G, 'A')
```
